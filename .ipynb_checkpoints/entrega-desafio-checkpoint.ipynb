# Introdução

O objetivo deste relatório é analisar o desempenho de quatro lojas com base em métricas comerciais e operacionais para auxiliar o Senhor João a tomar uma decisão estratégica sobre qual loja vender. Foram considerados aspectos como faturamento, avaliação dos clientes, frete médio, volume de vendas por categoria e desempenho por produto.

```python
# importações 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leitura do dataset (arquivos csv)
loja1 = pd.read_csv("loja_1.csv")
loja2 = pd.read_csv("loja_2.csv")
loja3 = pd.read_csv("loja_3.csv")
loja4 = pd.read_csv("loja_4.csv")
```

## Funções

```python
# FATURAMENTO 
# função para trazer o faturamento de cada loja;

def faturamento(dataset):
    faturamento_loja = sum(dataset["Preço"])
    return faturamento_loja
    #return f'{faturamento_loja:,.2f}' RETORNAR UMA STRING FORMATADA APENAS QUANDO FOR PRA EXIBIR
```

```python
# VENDAS POR CATEGORIA
# função que traz as vendas por categoria;

def vendas_categoria(dataset):
    resultado = dataset.groupby("Categoria do Produto").count()
    return resultado["Produto"].sort_values(ascending=False)
```

```python
# MEDIA DE AVALIACOES DAS LOJAS
# função que traz a media de avaliacoes das lojas;

def media_avaliacoes(dataset):
    media = dataset["Avaliação da compra"].mean()
    return round(media,2)
```

```python
# PRODUTOS MAIS VENDIDOS E MENOS VENDIDOS
# função que retorna os 5 produtos mais e menos vendidos

def mais_menos_vendidos(dataset, nome_loja):
    agrupados = dataset["Produto"].value_counts()
    
    mais_vendidos = agrupados.head(5).reset_index()
    mais_vendidos.columns = ["Produto", "Quantidade"]
    mais_vendidos["Tipo"] = "Mais Vendidos"
    mais_vendidos["Loja"] = nome_loja
    
    menos_vendidos = agrupados.tail(5).reset_index()
    menos_vendidos.columns = ["Produto", "Quantidade"]
    menos_vendidos["Tipo"] = "Menos Vendidos"
    menos_vendidos["Loja"] = nome_loja
    
    return pd.concat([mais_vendidos, menos_vendidos], ignore_index=True)
```

```python
# FRETE MÉDIO POR LOJA
# função que retorna o valor médio por Loja;

def media_frete(dataset):
    media = dataset["Frete"].mean()
    return round(media,2)
```

##  1. Análise do Faturamento

```python
print(f'Faturamento Loja_1: {faturamento(loja1):,.2f}')
print(f'Faturamento Loja_2: {faturamento(loja2):,.2f}')
print(f'Faturamento Loja_3: {faturamento(loja3):,.2f}')
print(f'Faturamento Loja_4: {faturamento(loja4):,.2f}')
```

## 2. Vendas por Categoria

```python
print(f'Vendas por categoria \n{vendas_categoria(loja1)}')
print(" ")
print(f'Vendas por categoria \n{vendas_categoria(loja2)}')
print(" ")
print(f'Vendas por categoria \n{vendas_categoria(loja3)}')
print(" ")
print(f'Vendas por categoria \n{vendas_categoria(loja4)}')
```

## 3. Média de Avaliação por Loja

```python
print(f'Média Avaliações Loja_1 {media_avaliacoes(loja1)}')
print(f'Média Avaliações Loja_2 {media_avaliacoes(loja2)}')
print(f'Média Avaliações Loja_3 {media_avaliacoes(loja3)}')
print(f'Média Avaliações Loja_4 {media_avaliacoes(loja4)}')
```

## 4. Produtos Mais e Menos Vendidos

```python
print(mais_menos_vendidos(loja1,"Loja 1"))
print(" ")
print(mais_menos_vendidos(loja2,"Loja 2"))
print(" ")
print(mais_menos_vendidos(loja3, "Loja 3"))
print(" ")
print(mais_menos_vendidos(loja4, "Loja 4"))
```

## 5. Frete Médio por Loja

```python
print(f'Frete Médio Loja_1 {media_frete(loja1)}')
print(f'Frete Médio Loja_2 {media_frete(loja2)}')
print(f'Frete Médio Loja_3 {media_frete(loja3)}')
print(f'Frete Médio Loja_4 {media_frete(loja4)}')
```

# Visualização

### Gráfico 1: Faturamento por Loja

```python
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
lista_faturamento = [faturamento(loja1), faturamento(loja2), faturamento(loja3), faturamento(loja4)]

plt.bar(lojas, lista_faturamento, color='skyblue') # criando o gráfico de barras
plt.title("Faturamento por Loja") # Titulo
plt.xlabel("Lojas") # Titulo eixo X
plt.ylabel("Faturamento (R$)") # Titulo eixo Y
plt.grid(axis='y', linestyle='--', alpha=0.5) # Linhas de grade

# colocando os valores no topo das barras
for i, valor in enumerate(lista_faturamento):
    plt.text(i, valor + max(lista_faturamento)*0.01, f'R${valor:,.2f}', ha='center')
```

### Gráfico 2: Vendas por Categoria

```python
# criando outros DF pra não comprometer o original
lj_1 = loja1
lj_2 = loja2
lj_3 = loja3
lj_4 = loja4

# adicionar uma coluna "loja" a cada dataset para melhor manipulação
lj_1["Loja"] = "Loja 1"
lj_2["Loja"] = "Loja 2"
lj_3["Loja"] = "Loja 3"
lj_4["Loja"] = "Loja 4"

#unindo em um único DF
juntos = pd.concat([lj_1, lj_2, lj_3, lj_4])

# .size() conta as linhas
# .count() conta tudo
vendas_por_categoria = juntos.groupby(["Categoria do Produto", "Loja"]).size().reset_index(name="Vendas")

df = vendas_por_categoria.reset_index()
df = df.drop("index", axis=1) 
df.columns = ["Categoria", "Loja", "Vendas"] # renomeando as colunas
```

```python
plt.figure(figsize=(12, 6)) # Tamanho do gráfico
grafico = sns.barplot(data=df, x="Categoria", y="Vendas", hue="Loja")

for p in grafico.patches:
    altura = p.get_height()
    x = p.get_x() + p.get_width() / 2
    y = altura
    grafico.annotate(
        f'{int(altura)}', 
        (x, y), 
        ha='center', 
        va='bottom', 
        fontsize=9,
        rotation=0  # direcao do texto
    )

plt.title("Vendas por Categoria")
plt.xticks(rotation=45) # Rotação da legenda no eixo X
plt.ylabel("Quantidade de Vendas")
plt.xlabel("Categoria do Produto")
plt.legend(title="Loja")
plt.tight_layout()
plt.show()
```

### Gráfico 3: Média de Avaliação por Loja

```python
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
lista_avaliacao = [media_avaliacoes(loja1), media_avaliacoes(loja2), media_avaliacoes(loja3), media_avaliacoes(loja4)]

plt.plot(lojas, lista_avaliacao, 'o', color='skyblue', markersize=12)
plt.title("Média de Avaliação por Loja")
plt.xlabel("Lojas")
plt.ylabel("Média")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
```

### Gráfico 4: Frete Médio por Loja

```python
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
lista_frete = [media_frete(loja1), media_frete(loja2), media_frete(loja3), media_frete(loja4)]

plt.pie(lista_frete, labels=lojas, autopct=lambda pct: f'R${pct/100*sum(lista_frete):.2f}')
plt.title("Média de Frete por Loja")
plt.tight_layout()
```

# Considerações

**FATURAMENTO:**

A **Loja 4** teve o menor faturamento, o que acende um alerta sobre sua performance.



**MÉDIA DE AVALIAÇÕES DOS CLIENTES:**

Embora a **Loja 1** tenha a menor avaliação, a **Loja 4** também não se destaca e tem um diferença sutil entre elas.

**FRETE MÉDIO**

A **Loja 4** possui o frete mais econômico, o que é um ponto positivo.

**CATEGORIAS MAIS E MENOS VENDIDAS**

As categorias móveis e eletrônicos lideram as vendas em todas as lojas.

Utilidades domésticas, livros e instrumentos musicais são as menos procuradas.

**PRODUTOS MAIS E MENOS VENDIDOS**


É importante observar que a **Loja 4** tem produtos relevantes entre os menos vendidos, como Geladeira e Guarda-roupas, o que pode indicar uma dificuldade maior em conversão de produtos de maior valor agregado.

## **Conclusão**

Baseado na análise dos dados, a recomendação é que a **Loja 4** seja vendida pelos seguintes motivos:

* Teve o menor faturamento.
* Não tem destaque nas avaliações dos clientes.
* Apesar de se destacar nos valores de fretes, as vendas não tem bom desempenho.
* Produtos importantes ficando entre os menos vendidos.

Então, com base na análise de desempenho das lojas vender a Loja 4 é a decisão mais estratégica.
